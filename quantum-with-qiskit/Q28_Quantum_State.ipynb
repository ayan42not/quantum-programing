_The overall probability must be 1 when we observe a quantum system._

For example, the following vectors <u>cannot</u> be a valid quantum state:

$$
    \myvector{ \dfrac{1}{2} \\ \dfrac{1}{2} }
    \mbox{ and }
    \myvector{ \dfrac{\sqrt{3}}{2} \\ \dfrac{1}{\sqrt{2}} }.
$$

For the first vector, the probabilities of observing the states $\ket{0} $ and $ \ket{1} $ are $ \dfrac{1}{4} $.

So, the overall probability of getting a result is $ \dfrac{1}{4} + \dfrac{1}{4} = \dfrac{1}{2} $, which is less than 1.

For the second vector, the probabilities of observing the states $\ket{0} $ and $ \ket{1} $ are respectively $ \dfrac{3}{4} $ and $ \dfrac{1}{2} $.

So, the overall probability of getting a result is $ \dfrac{3}{4} + \dfrac{1}{2} = \dfrac{5}{4} $, which is greater than 1.

1.  **The summation of amplitude squares must be 1 for a valid quantum state**.

2.  **More formally, a quantum state can be represented by a vector having length 1.**

3.  **The summation of amplitude squares gives the square of the length of vector.**

4.  **But, this summation is 1, and its square root is also 1. So, we can use the term length  in the definition.**("Uzunluğun karesi = 1" olduğu gibi, "uzunluk = 1" de doğrudur. Dolayısıyla tanımda "uzunluk" (length) terimini doğrudan kullanabiliriz. İkisi de aynı şeyi ifade eder çünkü 1'in karesi de 1'dir, karekökü de 1'dir.)

<i> <b>Technical notes:</b> We represent a quantum state as $ \ket{u} $ instead of $ u $. Remember the relation between the length and dot product: $ \norm{u} = \sqrt{\dot{u}{u}} $.
    
In quantum computation, we use inner product instead of dot product, which is defined on complex numbers. By using bra-ket notation, $ \norm{ \ket{u} } = \sqrt{ \braket{u}{u} } = 1 $, or equivalently $ \braket{u}{u} = 1 $, where $ \braket{u}{u} $ is a short form of $ \bra{u}\ket{u} $. For real-valued vectors, $ \braket{v}{v} = \dot{v}{v}  $.
</i>

<h3> Task 1 </h3>

If the following vectors are valid quantum states defined with real numbers, then what can be the values of $a$ and $b$?

$$
    \ket{v} = \myrvector{a \\ -0.1 \\ -0.3 \\ 0.4 \\ 0.5}
    ~~~~~ \mbox{and} ~~~~~
   \ket{u} = \myrvector{ \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{b}} \\ -\frac{1}{\sqrt{3}} }.
$$

```python
#
# your code is here
#    (you may find the values by hand (in mind) as well)
#
from math import sqrt

v=[-0.1,-0.3,0.4,0.5] # The a isn't written to intending that we're substracting the given probability  from the total probability.
u=[1/sqrt(1/2),-1/sqrt(3)]# The b isn't written to intending that we're substracting the given probability  from the total probability.

#If these vector represent a quantum state, as it is said us, it's inner product must be 1 ,  so does its lenght . In this etape we can find a and b

lenght_v=sum([v[i]*v[i] for i in range(len(v))])                    # or squares of lenght or amplitude which gives us the observing the probability of vector in the ket0 and ket1, respectively.
lenght_u=sum([u[i]*u[i] for i in range(len(u))])

a=sqrt(1-lenght_v)
b=1/(1-lenght_u)

print(f"a={a:.4f} and b={b:.4f}")


```

<h3> Quantum Operators </h3>

Once the quantum state is defined, the definition of quantum operator is very easy.
**Any length preserving (square) matrix is a quantum operator.**( Burada lenght preserving önceki lenght of bir state süperposizyonunda korunmalı ki bu bir sistemin operatör uygulanmadan önce statelere dağılmış olasılık dağımlarının toplamanın 1 olması gereğindendir.)

Remember Hadamard operator:

Randomly create a 2-dimensional quantum state, and test whether Hadamard operator preserves its length or not.

<b>Write a function</b> that returns a randomly created 2-dimensional quantum state.

<i>Hint:

    <li> Pick two random values between -100 and 100 for the amplitudes of state 0 and state 1 </li>
    <li> Find an appropriate normalization factor to divide each amplitude such that the length of quantum state should be 1 </li>

<b>Write a function</b> that determines whether a given vector is a valid quantum state or not.

(Due to precision problem, the summation of squares may not be exactly 1 but very close to 1, e.g., 0.9999999999999998.)

<b>Repeat 10 times:</b>
<ul>
    <li> Randomly pick a quantum state </li>
    <li> Check whether the picked quantum state is valid </li>
    <li> Multiply Hadamard operator with the randomly created quantum state </li>
    <li> Check whether the quantum state in result is valid </li>
</ul>

```python
#Normalleştirilmiş vektörün bileşenlerinin kareleri toplamı 1 olmalı, kendileri toplamı değil.Sonra her bir bileşeni bu norma bölünmeli.Bu normalize

#Bir vektör kendisiyle çarpılarak normu bulunur daha kendi bu boyuna bölünür yani her bir bileşeni!
import math
from random import randrange
def random_quantum_state():
    # quantum state
    quantum_state=[0,0]
    amp_0,amp_1=randrange(-100,100),randrange(-100,100)
    quantum_state[0]+=amp_0
    quantum_state[1]+=amp_1


    return quantum_state

hadamard=[[1/math.sqrt(2),1/math.sqrt(2)],[1/math.sqrt(2),-1/math.sqrt(2)]]
def Hdmrd_preser(v):
  norm_input = math.sqrt(v[0]**2 + v[1]**2)
  out_state=[0,0]
  for i in range(len(hadamard)):
    for j in range(len(hadamard)):
      out_state[i]+=hadamard[i][j]*v[j]

  norm_output= math.sqrt(out_state[0]**2 + out_state[1]**2)


  if  abs(norm_output -norm_input) < 1e-2:
    print(f"Hadamard applied on {v} then its lenght is preserved")
    print(f"It represents a quantum state!")

  else :
    print(f"Hadamard applied on {v} then its lenght not preserved")
    print(f"It does not a quantum state!")

for i in range(100):
  rdm_q_s=random_quantum_state()
  Hdmrd_preser(rdm_q_s)


```
